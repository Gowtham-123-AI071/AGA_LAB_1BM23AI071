```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import BernoulliRBM
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

print("Demonstrating Feature Extraction with Restriction Boltzman Machine(RBM)\n")
print("1.Generating sample data (digits datasets)....")
```

```python
digits=datasets.load_digits(n_class=6)
X_data=digits.data
y_data=digits.target

scaler=MinMaxScaler()
X_data=scaler.fit_transform(X_data)

X_train,X_test,y_train,y_test=train_test_split(X_data,y_data,test_size=0.2,random_state=42)

print(f"Original data shape:{X_train.shape}")
print(f"Number of features (visible units):{X_train.shape[1]}")
```

```python
print("2.Initialzing and training the BernoulliRBM...")
rbm=BernoulliRBM(n_components=64,learning_rate=0.01,n_iter=100,batch_size=10,random_state=0,verbose=True)

rbm.fit(X_train)
print("RBM training complete..\n")
```

```python
print("3.Extracting features from the training data...")
X_features=rbm.transform(X_train)
print(f"Original data shape:{X_train.shape}")
print(f"Extracted features shape:{X_features.shape}")
print(f"Each original sample of:{X_train.shape[1]} dimensions is now represented by{X_features.shape[1]} features")
```

```python
print("4.Demonstrating feature is represented by reconstructing the original data...")
x_constructed=rbm.gibbs(X_train)
print("original vs reconstructed sample(first 5 sample):")
```

```python
fig,axes=plt.subplots(nrows=2,ncols=5,figsize=(12,5))

for i in range(5):
    axes[0,i].imshow(X_train[i].reshape(8,8),cmap=plt.cm.gray_r)
    axes[0,i].set_title(f"original{i+1}")
    axes[0,i].axis('off')
    
    axes[1,i].imshow(X_train[i].reshape(8,8),cmap=plt.cm.gray_r)
    axes[1,i].set_title(f"original{i+1}")
    axes[1,i].axis('off')
    
plt.tight_layout()
plt.show()
```
